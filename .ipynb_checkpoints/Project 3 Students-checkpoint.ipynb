## 2. Data Preprocessing and Cleaning

```python
# drop datetime info
# -----------
chicago_inspections_2011_to_2013 = chicago_inspections_2011_to_2013.dropna().drop(['Inspection_Date',
                                                                                   'minDate',
                                                                                   'maxDate',
                                                                                   'Inspection_MonthYear'],
                                                                                  axis = 1)
```

```python
# process target 
# -----------
y = chicago_inspections_2011_to_2013['Results']

# decide if you want to binarize the outcome variable 
# -----------
# comment out the following lines of code if you don't want to binarize the target variable
y = y.replace({'Pass w/ Conditions': 'Pass'})
lb_style = LabelBinarizer()
y = lb_style.fit_transform(y)

# recode 0s and 1s so 1s are "Fail"
y = np.where(y == 1, 0 ,1)


# process features
# -----------

# create feature dataset
X = chicago_inspections_2011_to_2013.drop(columns = ['Results', 
                                                     'Facility_Type', 
                                                     'LICENSE_DESCRIPTION', 
                                                     'WARD_PRECINCT'])


# get dummies
X = pd.get_dummies(X)
```

### Model 1

Logistic regression models the probability of failure using a linear combination of features. The advantages of this model are high interpretability, fast training and prediction, and naturally calibrated probability outputs for ranking establishments. However, it cannot capture complex non-linear relationships or interactions without being adjusted and it may not be able to capture relationships that are highly non-linear. It's a good fit for this scenario because policymakers need to understand and justify why establishments are flagged for inspection, making transparency crucial, and food safety factors likely affect failure risk.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
param_grid_lr = {'C': [0.01, 0.1, 1.0, 10.0, 100.0],
                 'penalty': ['l2']}
model_lr = LogisticRegression(solver='liblinear', random_state=42)
grid_search_lr = GridSearchCV(estimator=model_lr,
                              param_grid=param_grid_lr,
                              scoring='recall',
                              cv=5,
                              verbose=1)
grid_search_lr.fit(X, y.ravel())
best_model_lr = grid_search_lr.best_estimator_
print(f"Best parameters for Logistic Regression: {grid_search_lr.best_params_}")
```

###  Model 2

Random Forest builds an ensemble of decision trees trained on bootstrapped data samples with random feature subsets at each split, and makes minimal distributional assumptions while automatically capturing non-linear relationships and feature interactions. The pros of the Random Forest model is that is robust to outliers, it can handle mixed data types without scaling, it has built-in feature importance metrics, and it reduces overfitting through ensemble averaging. The main downside is reduced interpretability compared to single models. This model is suited for food inspections because safety outcomes likely involve complex interactions (e.g., how do past violations interact with facility type) and Random Forests can capture these interactions automatically without requiring domain experts to manually specify these relationships.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
param_grid_rf = {'n_estimators': [50, 100, 200], 
                 'max_depth': [5, 10, 15],       
                 'min_samples_split': [2, 5]}   
model_rf = RandomForestClassifier(random_state=42)
grid_search_rf = GridSearchCV(estimator=model_rf,
                              param_grid=param_grid_rf,
                              scoring='recall',
                              cv=5,
                              verbose=1)
grid_search_rf.fit(X, y.ravel())
best_model_rf = grid_search_rf.best_estimator_
print(f"Best parameters for Random Forest: {grid_search_rf.best_params_}")
```

### Model 3

AdaBoost sequentially builds decision trees where each new tree corrects errors from previous trees using gradient descent optimization, incorporating sophisticated regularization techniques to prevent overfitting. It requires you to tune the hyperparameters and is less interpretable than simpler models. It's also less transparent than the logistic regression, but AdaBoost is more accurate

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
param_grid_ab = {'n_estimators': [50, 100, 200],
                 'learning_rate': [0.01, 0.1, 1.0]} 
model_ab = AdaBoostClassifier(random_state=42)
grid_search_ab = GridSearchCV(estimator=model_ab,
                              param_grid=param_grid_ab,
                              scoring='recall', 
                              cv=5, 
                              verbose=1)
grid_search_ab.fit(X, y.ravel())
best_model_ab = grid_search_ab.best_estimator_
print(f"Best parameters for AdaBoost: {grid_search_ab.best_params_}")
print(f"Best cross-validation score (recall): {grid_search_ab.best_score_:.4f}")
```

### Validation Metrics

Be sure to explain which of these metrics you would want to prioritize when conducting predictive auditing in this context and why.

**Hint**: Try writing a for loop to use [`cross_val_score()`](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html) to check for accuracy, precision, recall and f1 across all of your models.

```python
from sklearn.model_selection import cross_val_score
model_lr = LogisticRegression(solver='liblinear', random_state=42)
model_rf = RandomForestClassifier(random_state=42)
model_ab = AdaBoostClassifier(random_state=42)

models = [
    ("Logistic Regression", model_lr),
    ("Random Forest", model_rf),
    ("AdaBoost", model_ab)
]

scoring_metrics = ['accuracy', 'precision', 'recall', 'f1']

for name, model in models:
    print(f"\n--- Evaluating {name} ---")

    for scoring in scoring_metrics:
        scores = cross_val_score(model, X, y.ravel(), cv=5, scoring=scoring, n_jobs=-1)
        
        mean_score = scores.mean()
        std_score = scores.std()
        
        print(f"{scoring.capitalize()}: {mean_score:.4f} (95% CI: {mean_score - std_score * 2:.4f} to {mean_score + std_score * 2:.4f})")
```

## 5. Discussion Questions

1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?

**Metrics beyond accuracy are necessary in the social sciences and public policy because just using accuracy doesn't capture critical information about model performance, especially when dealing with imbalanced datasets. In the Chicago food inspection example, a model that predicts "Pass" for every establishment would achieve about 80% accuracy, but it would fail to identify any health violations. Failing to flag a restaurant with health violations (false negative) could result in illness or death, while incorrectly flagging a safe restaurant (false positive) wastes inspection resources. Accuracy treats both mistakes equally, even though they aren't. That's why it's important to use other metrics like recall (to catch violations), precision (to avoid wasting resources), and F1-score (to balance both).
Different stakeholders care about different metrics, and the choice of which to prioritize depends on the situation. For example, in public health, policymakers prioritize recall to protect citizens, whereas budget offices care about precision to save money. For food safety, a lower precision (some false alarms) is okay to maximize recall (catch actual violations) because the cost of missing a health hazard far outweighs conducting an unnecessary inspection. By reporting multiple metrics and examining the confusion matrix, we empower policymakers to understand the tradeoffs and make better policy decisions.**

2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decisionmaking and real world behavior?

**If establishments start adjusting their behavior to avoid detection, policymakers should adjust the models. Policymakers should regularly update models with new data to adapt to evolving behaviors and use ensemble methods that rotate between different feature sets to create a moving target. They should also prioritize hard-to-manipulate features like location, historical patterns, and neighborhood-level data. Continuing random inspection also prevents establishments from feeling safe and provides unbiased data to detect gaming, and monitoring for suspicious patterns helps identify manipulation attempts.**
